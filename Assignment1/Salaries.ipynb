```python
sns.heatmap(sal.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
sal['BasePay'] = sal['BasePay'].replace('Not Provided', np.nan).astype(float)
mean_value = sal['BasePay'].mean()
sal['BasePay'].fillna(mean_value, inplace=True)
```

```python
sns.histplot(sal['BasePay'], kde=True)
plt.title('Distribution of Base Pay')
plt.show()
```

```python
sal['TotalPay'].describe()
```

```python
sns.boxplot(x=sal['BasePay'])
plt.title('Box Plot of Base Pay')
plt.show()
```

```python
sns.lineplot(x=sal['Year'], y=sal['TotalPayBenefits'])
plt.title('Total Pay Benefits Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Pay Benefits')
plt.show()
```

```python
sns.histplot(data=sal, x='BasePay', hue='Year', kde=True, multiple='stack')
plt.title('Base Pay Distribution by Year')
plt.show()
```

```python
sns.countplot(x='JobTitle', data=sal, order=sal['JobTitle'].value_counts().head(5).index)
plt.title('Top 5 Job Titles')
plt.show()
```
